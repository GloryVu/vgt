```python
import tensorflow as tf

def extract_metrics_from_event_file(file_path):
    # Initialize lists to hold the data
    step_numbers = []
    metrics = {}
    
    # Iterate through the event file
    for event in tf.compat.v1.train.summary_iterator(file_path):
        for value in event.summary.value:
            if value.tag not in metrics:
                metrics[value.tag] = []
            metrics[value.tag].append(value.simple_value)
            step_numbers.append(event.step)
    
    return step_numbers, metrics

# Example usage
file_path = 'events.out.tfevents.1717612161.aies-gpu-a30.1207434.0'
steps, extracted_metrics = extract_metrics_from_event_file(file_path)

# Print extracted metrics
print("Steps:", steps)
for metric_name, metric_values in extracted_metrics.items():
    print(f"{metric_name}: {metric_values}")
```

```python
import json
import os
train_metics ={}
for model in sorted(os.listdir('./eval_9k/')):
    with open(f'./eval_9k/{model}') as f:
        metrics = json.load(f)
        for metric in metrics['bbox'].keys():
            if f'bbox/{metric}' in train_metics.keys():
                train_metics[f'bbox/{metric}'].append(metrics['bbox'][metric])
            else:
                train_metics[f'bbox/{metric}']=[metrics['bbox'][metric]]
```

```python
import tensorflow as tf
with tf.summary.create_file_writer('train').as_default():
    # Write each metric to the event file
    # step = [9000,12000,21000,24000,26000,30000,35000]
    for name, values in train_metics.items():
        for i,value in enumerate(values):
            tf.summary.scalar(name, value, (i+1)*3600)  # Assuming step=0 for simplicity

print("Metrics have been written to event.out.tfevent file.")
```

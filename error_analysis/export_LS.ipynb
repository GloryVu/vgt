```python
ls_annotation =[]
for img in annotations['images'] :
    if img['id'] in img_ids:
        img = img['file_name']#.replace('.png','.jpg')
        
        ls_annotation.append({
                "data": {
                "image": f"/data/local-files/?d=documents/{img}"
            }
        })
```

```python
idx_img_mapping = {annotation['data']['image'].replace('/data/local-files/?d=documents/',''): i for i,annotation in enumerate(ls_annotation)}
```

```python
for file_name, errors in df.groupby("file_name"):
    file_name=file_name#.replace('.png','.jpg')
    if file_name not in idx_img_mapping:
        continue
    # print(errors['error_percent'].)
    errors['cls_loc']=0
    error_json = {error_type:errors[error_type].values[0] for error_type in data_fields}
    for key in error_json.keys():
        if key != 'error_percent':
            error_json[key] = int(error_json[key])
        else:
            error_json[key] = round(error_json[key],4)
    
    ls_annotation[idx_img_mapping[file_name]]['data']= {
        **ls_annotation[idx_img_mapping[file_name]]['data'],
        **error_json
        }
```

```python
def convert_coco_to_ls(error_gt_annotations,error_pred_annotations, original_gt_annotations,original_pred_annotations, file_name):
    error_gt_ls_annotations = []
    error_pred_ls_annotations = []
    merge_ls_annotations = []
    original_gt_ls_annotations = []
    original_pred_ls_annotations = []
    file_name = file_name#.replace('.jpg','.png')
    gt_annotations_file = []
    def calculate_iou(bbox1, bbox2):
    # Convert [xmin, ymin, w, h] to [xmin, ymin, xmax, ymax]
        xmin1, ymin1, w1, h1 = bbox1
        xmin2, ymin2, w2, h2 = bbox2
        
        xmax1, ymax1 = xmin1 + w1, ymin1 + h1
        xmax2, ymax2 = xmin2 + w2, ymin2 + h2
        
        # Calculate intersection coordinates
        x_inter_min = max(xmin1, xmin2)
        y_inter_min = max(ymin1, ymin2)
        x_inter_max = min(xmax1, xmax2)
        y_inter_max = min(ymax1, ymax2)
        
        # Calculate intersection area
        if x_inter_max > x_inter_min and y_inter_max > y_inter_min:
            inter_area = (x_inter_max - x_inter_min) * (y_inter_max - y_inter_min)
        else:
            inter_area = 0
        
        # Calculate areas of each bbox
        area1 = w1 * h1
        area2 = w2 * h2
        
        # Calculate union area
        union_area = area1 + area2 - inter_area
        
        # Calculate IoU
        iou = inter_area / union_area if union_area != 0 else 0
        
        return iou
    def find_matches_bbox(bbox):
        max_iou = 0
        for annot in gt_annotations_file:
            max_iou = max(max_iou,calculate_iou(bbox,annot['bbox']))
        return max_iou
    
    for annotation in error_gt_annotations:
        if annotation['image_id'] == gt_id_img_mapping[file_name]['id']:
            gt_annotations_file.append(annotation)
            ow,oh= gt_id_img_mapping[file_name]['width'],gt_id_img_mapping[file_name]['height']
            x, y, w, h = annotation['bbox']
            
            error_gt_ls_annotations.append(
                {
                'type': 'rectanglelabels',
                'value': {'x': x/ow*100,
                'y': y/oh*100,
                'width': w/ow*100,
                'height': h/oh*100,
                'rotation': 0,
                'rectanglelabels': [label_mapping[annotation['category_id']]]},
                'to_name': 'image',
                'from_name': 'label',
                'image_rotation': 0,
                'original_width': ow,
                'original_height': oh,
                }
            )
            merge_ls_annotations.append(
                {
                'type': 'rectanglelabels',
                'value': {'x': x/ow*100,
                'y': y/oh*100,
                'width': w/ow*100,
                'height': h/oh*100,
                'rotation': 0,
                'rectanglelabels': ['gt_'+label_mapping[annotation['category_id']]]},
                'to_name': 'image',
                'from_name': 'label',
                'image_rotation': 0,
                'original_width': ow,
                'original_height': oh,
                }
            )
    for annotation in error_pred_annotations:
        if annotation['image_id'] == gt_id_img_mapping[file_name]['id'] and annotation['score'] >= 0.5:
            ow,oh= gt_id_img_mapping[file_name]['width'],gt_id_img_mapping[file_name]['height']
            x, y, w, h = annotation['bbox']
            
            error_pred_ls_annotations.append(
                {
                'type': 'rectanglelabels',
                'value': {'x': x/ow*100,
                'y': y/oh*100,
                'width': w/ow*100,
                'height': h/oh*100,
                'rotation': 0,
                'rectanglelabels': ['pred_'+label_mapping[annotation['category_id']]]},
                'to_name': 'image',
                'from_name': 'label',
                'score': float(annotation['score']),#find_matches_bbox(annotation['bbox']),
                'image_rotation': 0,
                'original_width': ow,
                'original_height': oh,
                'meta':{
                    "text": [
                               str(find_matches_bbox(annotation['bbox']))
                            ]
                }
                }
            )
            merge_ls_annotations.append(
                {
                'type': 'rectanglelabels',
                'value': {'x': x/ow*100,
                'y': y/oh*100,
                'width': w/ow*100,
                'height': h/oh*100,
                'rotation': 0,
                'rectanglelabels': ['pred_'+label_mapping[annotation['category_id']]]},
                'to_name': 'image',
                'from_name': 'label',
                'score': float(annotation['score']),
                'image_rotation': 0,
                'original_width': ow,
                'original_height': oh,
                'meta':{
                    "text": [
                               str(find_matches_bbox(annotation['bbox']))
                            ]
                }
                }
            )
    for annotation in original_pred_annotations:
        if annotation['image_id'] == gt_id_img_mapping[file_name]['id'] and annotation['score'] >= 0.5:
            ow,oh= gt_id_img_mapping[file_name]['width'],gt_id_img_mapping[file_name]['height']
            x, y, w, h = annotation['bbox']
            original_pred_ls_annotations.append(
                {
                'type': 'rectanglelabels',
                'value': {'x': x/ow*100,
                'y': y/oh*100,
                'width': w/ow*100,
                'height': h/oh*100,
                'rotation': 0,
                'rectanglelabels': ['pred_'+label_mapping[annotation['category_id']]]},
                'to_name': 'image',
                'from_name': 'label',
                'score': float(annotation['score']),
                'image_rotation': 0,
                'original_width': ow,
                'original_height': oh,
                }
            )
    for annotation in original_gt_annotations:
        if annotation['image_id'] == gt_id_img_mapping[file_name]['id']:
            ow,oh= gt_id_img_mapping[file_name]['width'],gt_id_img_mapping[file_name]['height']
            x, y, w, h = annotation['bbox']
            original_gt_ls_annotations.append(
                {
                'type': 'rectanglelabels',
                'value': {'x': x/ow*100,
                'y': y/oh*100,
                'width': w/ow*100,
                'height': h/oh*100,
                'rotation': 0,
                'rectanglelabels': ['gt_'+label_mapping[annotation['category_id']]]},
                'to_name': 'image',
                'from_name': 'label',
                # 'score': float(annotation['score']),
                'image_rotation': 0,
                'original_width': ow,
                'original_height': oh,
                }
            ) 
    return [
        { 'result': error_gt_ls_annotations,
            "ground_truth": False,
            },
        { 'result': original_gt_ls_annotations,
            "ground_truth": False,
            },
        # { 'result': pred_ls_annotations,
        #     "ground_truth": False,
        #     },
        
        { 'result': original_pred_ls_annotations,
            'model_version': 'original'
        },
        { 'result': error_pred_ls_annotations,
            'model_version': 'error'
        },
        { 'result': merge_ls_annotations,
            "model_version": 'merge',
        },
    ]
        
```

```python
for annot in ls_annotation:
    file_name = annot['data']['image'].replace('/data/local-files/?d=documents/','')
    if 'error_percent' not in annot['data']:
        annot['data'] = {**annot['data'],
                         'error_percent': 0,
                        'correct': 0,
                        'background': 0,
                        'localization': 0,
                        'classification': 0,
                        'missed': 0,
                        'cls_loc': 0
                        }
    annot['data']['set']=dataset
    converted_ls = convert_coco_to_ls(gt_fail,error_pred_labels,gt_labels['annotations'],pred_labels,file_name)
    annot['data']['classes_fail'] =','.join(set([annot['value']['rectanglelabels'][0] for annot in converted_ls[-1]['result']]))
    annot['annotations'] = converted_ls[:2]
    annot['predictions'] = converted_ls[2:]
    # annot['predictions'] = [converted_ls[-1]]
```
